## Intro

The purpose of this project is to create a Juypter Notebook that utilizes Polars to generate descriptive statistics. It establishes an environment on codespaces and uses Github Actions to run a Makefile for : make install, make test, make format, and make lint. It loads the Top Wealthiest data set from Kaggler and utiize a library to run functions that will generate a pie chart as well as some summary statistics about the data.

```python
from mylib.lib import (
    load_data,
    plot_pie_chart,
    grab_mean,
    grab_min,
    grab_std,
    grab_max,
)


csv_data = "Top_1000_wealthiest_people.csv"
```

```python
general_df = load_data(csv_data)
general_df.head()
assert general_df is not None
assert general_df.shape == (1000, 5)
```

```python
# Calculate statistics
mean_value = grab_mean(general_df, "Net Worth (in billions)")
min_value = grab_min(general_df, "Net Worth (in billions)")
std_value = grab_std(general_df, "Net Worth (in billions)")
max_value = grab_max(general_df, "Net Worth (in billions)")

# Expected values
test_mean = 102.6
test_std = 56.8
test_max = 199.6
test_min = 1.5

# Precision tolerance
tolerance = 0.5

# Use assertAlmostEqual for floating-point comparisons
assert (
    abs(mean_value - test_mean) < tolerance
), f"Mean value {mean_value} differs from expected {test_mean}"
assert (
    abs(std_value - test_std) < tolerance
), f"Std value {std_value} differs from expected {test_std}"
assert (
    abs(max_value - test_max) < tolerance
), f"Max value {max_value} differs from expected {test_max}"
assert (
    abs(min_value - test_min) < tolerance
), f"Min value {min_value} differs from expected {test_min}"
```
